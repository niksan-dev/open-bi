```python
from openbi.datasource.json_connector import JSONConnector
from openbi.datasource.rest_connector import RESTConnector

connector = JSONConnector(

    source="../examples/sales.json",

    dataset_name="Retail Sales"

)

result = connector.load()
print(result.statistics)
dataset = result.dataset
print(dataset)
print(dataset.name)
print(dataset.source)



connector = RESTConnector(

    "https://jsonplaceholder.typicode.com/users"

)

result = connector.load()

dataset = result.dataset
print("Rest Dataset::" , dataset)
```
