```python
import pandas as pd

from openbi.model.dataset import Dataset
from openbi.model.table import Table
from openbi.pipeline.pipeline_builder import DefaultPipeline

# ---------------------------------------
# Create Dataset
# ---------------------------------------

dataset = Dataset(name="Retail Sales")

# ---------------------------------------
# Load Excel
# ---------------------------------------

df = pd.read_excel("../examples/sales.xlsx")

print("Original Data")
display(df)

# ---------------------------------------
# Execute Pipeline
# ---------------------------------------

pipeline = DefaultPipeline.create()

result = pipeline.execute(df)

stats = result.statistics

clean_df = result.dataframe



print("Cleaned Data")
display(clean_df)

# ---------------------------------------
# Create Table
# ---------------------------------------

orders = Table.from_dataframe(
    name="Orders",
    dataframe=clean_df
)

dataset.model.add_table(orders)

# ---------------------------------------
# Verify
# ---------------------------------------



for step in stats.steps:

    print(step.step_name)

    print("Rows Before :", step.rows_before)

    print("Rows After  :", step.rows_after)

    print("Changed     :", step.values_changed)

    print("Time (ms)   :", step.execution_time_ms)

    print("-" * 40)

print()
print("Pipeline Statistics")
print("-------------------")

print("Rows Before :", stats.rows_before)
print("Rows After  :", stats.rows_after)

print("Columns     :", stats.columns_after)

print("Execution   :", stats.total_execution_time_ms)

```
